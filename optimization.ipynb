```python
# Visualize gradient descent of quadratic function
%matplotlib inline
import itertools
import numpy as np
import matplotlib.pyplot as plt
from optimization import gradient_descent

plt.xlabel('x')
plt.ylabel('cost')
xs = np.arange(-1, 1, 0.01)
plt.plot(xs, xs**2)
plt.show()

iterator = gradient_descent(lambda x: x * x, lambda x: 2.0 * x, 10.0, 0.9)
xs, costs = zip(*itertools.islice(iterator, 100))

plt.xlabel('step')
plt.ylabel('cost')
plt.plot(costs)
plt.show()

plt.xlabel('step')
plt.ylabel('x')
plt.plot(xs)
plt.show()
```
