```python
# Plot dicision boundary of logistic regression
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from classification import *

X_train = np.random.rand(10, 2)
y_train = X_train @ np.array([1, -1]) > 0

model = LogisticRegression(X_train, y_train)

X_test = np.random.rand(1000, 2)
prediction = model.predict(X_test)
plt.scatter(X_test[:,0], X_test[:,1], c=['r' if p else 'b' for p in prediction])
plt.show()
```

```python
# Plot dicision boundary of support vector machine
X_train = np.random.rand(10, 2)
y_train = np.sign(np.random.rand(10) - 0.5)

model = SupportVectorMachine(X_train, y_train, lambda x, y: np.exp(-np.sum((x - y)**2)*1000))

X_test = np.random.rand(1000, 2)
prediction = model.predict(X_test)

plt.scatter(X_test[:,0], X_test[:,1], c=['r' if p==1 else 'b' for p in prediction])
plt.show()
```
