```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
# Using open implementation of RBM
# https://gist.github.com/blackecho/db85fab069bd2d6fb3e7
from rbm_after_refactor import *
rbm = RBM(784, 10, visible_unit_type='gauss')
rbm.fit(mnist.train.images)
W = rbm.get_model_parameters()['W']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

_, a = plt.subplots(1, 10, figsize=(50, 10))
for i in range(10):
    image = W[:,i].reshape((28, 28))
    a[i].imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```
