```python
# Visualize linear regression
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from linear_regression import *

X_train = np.random.rand(10)
y_train = X_train + (np.random.rand(10) - 0.5)
X_train = X_train.reshape(10, 1)
model = LinearRegression(X_train, y_train)

plt.xlabel('x')
plt.ylabel('y')
plt.plot(X_train, y_train, 'bo')

X = np.arange(0, 1, 0.01).reshape(-1, 1)
y = model.predict(X)

plt.plot(X, y, 'r-')

X_test = np.random.rand(10)
y_test = X_test + (np.random.rand(10) - 0.5)
X_test = X_test.reshape(10, 1)

plt.plot(X_test, y_test, 'go')
plt.show()

print('Red line: linear regression')
print('Blue points: training data')
print('Green points: test data')
print('Mean squared error', model.error(X_test, y_test))
```

```python
# Visualize linear regression with various dgree of polynomial
X_train = np.random.rand(10)
y_train = X_train ** 3 + np.random.rand(10) / 10
X_train = X_train.reshape(10, 1)

X_test = np.random.rand(10)
y_test = X_test ** 3 + np.random.rand(10) / 10
X_test = X_test.reshape(10, 1)

plt.xlabel('x')
plt.ylabel('y')
plt.plot(X_train, y_train, 'bo')
plt.plot(X_test, y_test, 'go')
colors = {1:'r', 5:'g', 9:'b'}

train_errors = []
test_errors = []

for degree in range(1, 10):
    model = LinearRegression(X_train, y_train, False, degree)
    train_errors.append(model.error(X_train, y_train))
    test_errors.append(model.error(X_test, y_test))
    if degree in colors:
        X = np.arange(0, 1, 0.01).reshape(-1, 1)
        y = model.predict(X)
        plt.plot(X, y, colors[degree])

plt.show()
print('Red line: polynomial of 1')
print('Green curve: polynomial of 5')
print('Blue curve: polynomial of 9')


plt.xlabel('polynomial degree')
plt.ylabel('mean squared error')
plt.plot(range(1, 10), train_errors, 'b')
plt.plot(range(1, 10), test_errors, 'r')
plt.show()

print('Red line: test error')
print('Blue line: training error')
```
